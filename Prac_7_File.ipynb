```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parameters
n = 10  # number of trials
p = 0.35  # probability of success

# List of possible outcomes (0 to n)
r = list(range(n + 1))

# Calculate the binomial PMF
return_val = binom.pmf(r, n, p)

# Loop through each value and print the PMF value
for i in range(n + 1):
    print("Probability of {i} successes: {return_val[i]}")
```

```python
from numpy.random import normal
from numpy import mean, std
from scipy.stats import norm

# Generate a sample of size 100 from a normal distribution with mean = 50 and std = 5
sample = normal(loc=50, scale=5, size=100)

# Calculate the mean and standard deviation of the sample
sample_mean = mean(sample)
sample_std = std(sample)

# Print the mean and standard deviation
print("Mean: %.3f, Standard Deviation: %.3f" % (sample_mean, sample_std))

# Create a normal distribution object using the calculated mean and standard deviation
dist = norm(sample_mean, sample_std)

# Define a range of values to calculate the PDF
values = [value for value in range(30, 70)]

# Calculate the probability density function (PDF) for each value in the range
probs = [dist.pdf(value) for value in values]

# Display the PDF values
probs
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Number of data points
n = 500

# Generate random data from a standard normal distribution
data = np.random.randn(n)

# Create a histogram of the data with 10 bins
count, bins_count = np.histogram(data, bins=10)

# Calculate the PDF (Probability Density Function)
pdf = count / sum(count)

# Calculate the CDF (Cumulative Distribution Function)
cdf = np.cumsum(pdf)

# Plot the PDF
plt.plot(bins_count[1:], pdf, color="red", label="PDF")

# Plot the CDF
plt.plot(bins_count[1:], cdf, label="CDF")

# Add the legend to the plot
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Define the discrete random variable
xk = np.arange(7)
pk = (0.1, 0.2, 0.3, 0.1, 0.1, 0.0, 0.2)
custm = stats.rv_discrete(name="custm", values=(xk, pk))

# Create the plot
fig, ax = plt.subplots(1, 1)
ax.plot(xk, custm.pmf(xk), "ro", ms=12, mec="r")
ax.vlines(xk, 0, custm.pmf(xk), colors='r', lw=4)

plt.title("Probability Mass Function")
plt.xlabel("x")
plt.ylabel("Probability")
plt.grid()
plt.show()
```
